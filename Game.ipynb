```python
import random
import time
def game() :
  Player1_HP = 100
  Player2_HP = 100
  Player1_Defence = False
  Player2_Defence = False
  turn = 1
  while (Player1_HP>0 and Player2_HP>0):
    print("\n-------------Turn Start---------------")
    print("Player 1s HP :", Player1_HP)
    print("Player 2s HP :", Player2_HP)

    if turn == 1 :
      print("player 1's turn")
      choice = input("Press Enter to attack for attack and defence for defence :")
      damage = random.randint(10,20)
      if choice == "attack":
        print("Generating Attack ............💥")
        #time.sleep(2)
        if Player2_Defence == True :
          damage = damage//2
          Player2_Defence == False
        Player2_HP = Player2_HP - damage
        print("Player1 has damaged Player 2 by ", damage ,"damage⚔️")
      else :
        print("Player 1 use Defence")
        Player1_Defence = True
      turn = 2
    else :
      print("player 2's turn")
      choice = Computer_choice(Player1_HP=Player1_HP,Player2_HP=Player2_HP,Player1_Defence=Player1_Defence,Player2_Defence=Player2_Defence)
      damage = random.randint(10,20)
      if choice == "attack" :
        print("Generating Attack ............💥")
        #time.sleep(2)
        if Player1_Defence == True :
            damage = damage//2
            Player1_Defence == False
        Player1_HP = Player1_HP - damage
        print("Player2 has damaged Player 1 by ", damage ,"damage⚔️")
      else :
        print("Player 1 use Defence")
        Player2_Defence = True
      turn = 1
  if Player1_HP>0 :
    #print("Player1 is winner")
    return "Player1"
  else :
    #print("Game over!")
    return "Player2"
```

```python
def Computer_choice(Player1_HP,Player2_HP,Player1_Defence, Player2_Defence):
  if(Player1_Defence):
    return "attack"
  elif(Player2_Defence):
    return "attack"
  elif(Player1_HP<30):
    return "attack"
  elif(Player2_HP<20):
    return "defence"
  elif random.random()<0.7 :
    return "attack"
```

```python
num_games = 1
player1_wins = 0
player2_wins = 0
for i in range(num_games):
  winner = game()
  if winner == "Player1" :
    player1_wins = player1_wins+1
  else :
    player2_wins = player2_wins+1
print(player1_wins,player2_wins)

import matplotlib.pyplot as plt
plt.figure(figsize=(5,5))
x_axis = ['Player 1','Player 2']
y_axis = [player1_wins,player2_wins]

plt.bar(x_axis,y_axis)
plt.ylabel("No of Wins")
plt.title("Game Results")
```
